```python
import pandas as pd
import os

df = pd.read_csv(os.path.join('data','casos-br-total.csv'), header=None)
df.columns = df.iloc[0]
df = df.iloc[1:]

df[['Suspeitos', 'Confirmados', 'Morte']] = df[['Suspeitos', 'Confirmados', 'Morte']].apply(pd.to_numeric)

df.head(len(df))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

plt.figure(figsize=(10, 5))
plt.plot(df['Data'], df['Suspeitos'].to_numpy(), linewidth=3, color='coral', label=None)
plt.xticks(df['Data'], rotation=90, fontsize=14)
plt.yticks(fontsize=14)
plt.title('Número de Casos Suspeitos', fontsize=21)

plt.axvline(3, 0, 8000, color='dodgerblue', linestyle='--', linewidth=2, alpha=0.7, label='Sequenciamento do Genoma')
plt.axvline(9, 0, 8000, color='k', linestyle='dotted', linewidth=2, alpha=1, label='Transmissão Comunitária')

plt.legend(loc=2, fancybox=True, fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Data'], df['Confirmados'].to_numpy(), linewidth=3, color='red', label=None)
plt.xticks(df['Data'], rotation=90, fontsize=14)
plt.yticks(fontsize=14)
plt.title('Número de Casos Confirmados', fontsize=21)

plt.axvline(3, 0, 8000, color='dodgerblue', linestyle='--', linewidth=2, alpha=0.7, label='Sequenciamento do Genoma')
plt.axvline(9, 0, 8000, color='k', linestyle='dotted', linewidth=2, alpha=1, label='Transmissão Comunitária')

plt.legend(loc=2, fancybox=True, fontsize=12)
plt.show()
```
